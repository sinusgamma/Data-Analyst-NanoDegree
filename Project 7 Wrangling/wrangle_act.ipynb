### Cleaning Tidiness

#### Tidiness 1

Define: Merge tables, we need only one table for the tweet data.

Code:

```python
# I don't use copies of original dataframes, because my first step is merging the dataframes.
# I could copy a dataframe to a new one with the following code:
# df_copy = df.copy()
```

```python
# merge tables on id or tweet_id
clean_df = pd.merge(twitter_archive_df, prediction_df, left_on ='tweet_id', right_on ='tweet_id', how= 'left' )
clean_df = pd.merge(clean_df, json_df, left_on ='tweet_id', right_on ='id', how= 'left' )
```

Test:

```python
list(clean_df)
```

#### Tidiness 2

Define: Transform the dog type columns to one column
    
Code:

```python
# delete "None" string
clean_df['doggo'].replace('None', '', inplace=True)
clean_df['floofer'].replace('None', '', inplace=True)
clean_df['pupper'].replace('None', '', inplace=True)
clean_df['puppo'].replace('None', '', inplace=True)
```

```python
# make the dog_type column from the distinct columns
clean_df['dog_type'] = (clean_df['doggo'] + 
                            clean_df['floofer'] + 
                            clean_df['pupper'] + 
                            clean_df['puppo'])
```

Test:

```python
clean_df['dog_type'].value_counts()
```

In this column we don't have 'None' strings, but we have '' or empty strings, wich isn't better. We handle this in the quality part of the cleaning.

### Visualize Data

```python
# build dataframe with dog breeds and breed counts
breed_counts_df = clean_df['p1'].value_counts()[:20].rename_axis('unique_values').reset_index(name='counts')
```

```python
# setup the barplot
f, ax = plt.subplots(figsize=(10, 10))
ax = sns.barplot(x="counts", y="unique_values", data=breed_counts_df, color="salmon")
ax.set_xlabel("number of p1 predictions")
ax.set_ylabel("dog breed")
ax.set_title("Dog Breed Counts by the most likely predictions")
```

In the above image we can see the the most popular dog breeds by occurrence according to the image predictions p1 (most probable) guess.

```python
f, ax = plt.subplots(figsize=(10, 10))
ax.set_yscale('log')
ax.set_xscale('log')
plt.scatter(clean_df["retweet_count"], clean_df["favorite_count"], alpha=0.3)
ax.set_xlabel("retweet count")
ax.set_ylabel("favorite count")
ax.set_title("Scatterplot of retweet count and favorite count with log scale")
```

In the plot above we can see that as the larger the retweet count of a dog the larger the favorite count most of the time.
